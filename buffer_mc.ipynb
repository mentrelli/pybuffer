```python
import numpy as np
import matplotlib.pyplot as plt
%pylab inline
```

```python
N, xmax = 100, 10

xx = np.ndarray(N+2)
pp = np.zeros(N+2)
for i in range(N+2):
    xx[i] = i*(1.0*xmax)/N

Ni = 100000
dp = 1.0/Ni
for i in range(Ni-1):
    x = np.random.exponential(1)
    idx = int(N*x/xmax)
    if idx > N+1: idx = N+1
    pp[idx] += dp

plt.plot(xx,pp,'.-')
```
